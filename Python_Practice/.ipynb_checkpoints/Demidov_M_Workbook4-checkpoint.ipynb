## 1.1.1 –ü—Ä–∏–º–µ—Ä
–ü—Ä–æ–≤–µ–¥–µ–º –ø—Ä—è–º—É—é ùë¶ = ùëöùë• + ùëè —á–µ—Ä–µ–∑ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞–ª—å–Ω—ã–µ —Ç–æ—á–∫–∏.

```python
import numpy as np
x = np.array([0, 1, 2, 3])
y = np.array([-1, 0.2, 0.9, 2.1])

#–ü–µ—Ä–µ–ø–∏—à–µ–º –ª–∏–Ω–µ–π–Ω–æ–µ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ y = mx + c –∫–∞–∫ y = Ap, –≥–¥–µ A = [[x 1]] –∏ p = [[m], [c]]
#–ü–æ—Å—Ç—Ä–æ–∏–º A –ø–æ x:

A = np.vstack([x, np.ones(len(x))]).T
print(A)

#–ò—Å–ø–æ–ª—å–∑—É–µ–º –º–µ—Ç–æ–¥ Lstsq –¥–ª—è —Ä–µ—à–µ–Ω–∏—è –µ–≥–æ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ –≤–µ–∫—Ç–æ—Ä–∞ p.
m, c = np.linalg.lstsq(A,  y, rcond = None)[0]
print(m, c)

#–ü–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –ø–æ–ª—É—á–µ–Ω–Ω–æ–π –ø—Ä—è–º–æ–π –∏ —É–∫–∞–∂–µ–º –Ω–∞ –Ω–µ–º —Ç–æ—á–∫–∏.
import matplotlib.pyplot as plt
plt.plot(x, y, 'o', label='–∏—Å—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ', markersize=10)
plt.plot(x, m*x+c, 'r', label='–õ–∏–Ω–µ–π–Ω–∞—è —ç–∫—Å—Ç—Ä–∞–ø–æ–ª—è—Ü–∏—è', markersize=10)
plt.legend()
plt.show()
```

## 1.1.2 –ü—Ä–∏–º–µ—Ä
–ü—É—Å—Ç—å ùë•, ùë¶ ‚Äì –≤–µ–∫—Ç–æ—Ä–∞ –¥–ª–∏–Ω–æ–π ùëõ > 3 (—Ç–æ—á–µ–∫ > 3). –ó–∞–¥–∞—á–∞ –∑–∞–∫–ª—é—á–∞–µ—Ç—Å—è –≤
–ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–∏ —ç—Å—Ç—Ä–∞–ø–æ–ª—è—Ü–∏–æ–Ω–Ω–æ–≥–æ –ø–æ–ª–∏–Ω–æ–º–∞ –≤—Ç–æ—Ä–æ–≥–æ –ø–æ—Ä—è–¥–∫–∞ (–ø–∞—Ä–∞–±–æ–ª—ã). –¢–∞–∫–∏–º
–æ–±—Ä–∞–∑–æ–º, –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –Ω–∞–π—Ç–∏ —Ç–∞–∫–∏–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –ø–æ—Ä–∏–Ω–æ–º–∞ ùëé, ùëè, ùëê –ø–æ –º–µ—Ç–æ–¥—É
–Ω–∞–∏–º–µ–Ω—å—à–∏—Ö –∫–≤–∞–¥—Ä–∞—Ç–æ–≤. –î–∞–Ω–Ω—ã–µ –º—Ç–æ–≥—É—Ç –±—ã—Ç—å –ø–æ–ª—É—á–µ–Ω—ã –≤ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–µ
–∏–∑–º–µ—Ä–µ–Ω–∏–π. –ü–æ–∫–∞–∂–µ–º –ø—Ä–∏–º–µ—Ä –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏ –¥–∞–Ω–Ω—ã—Ö —Å–ª—É—á–∞–π–Ω—ã–º –æ–±—Ä–∞–∑–æ–º –∏
–∑–∞–≥—Ä—É–∑–∫–∏ –∏—Ö –∏–∑ —Ñ–∞–π–ª–∞.

```python
from numpy import *
from numpy.random import *
#–≥–µ–Ω–µ—Ä–∏—Ä—É–µ–º —Å–ª—É—á–∞–π–Ω—ã–µ x –∏ y
delta = 1.0
x = linspace(-5, 5, 11)
y = x**2+delta*(rand(11)-0.5)
x += delta*(rand(11)-0.5)

#–ó–∞–ø–∏—Å—ã–≤–∞–µ–º –¥–∞–Ω–Ω—ã–µ –≤ —Ñ–∞–π–ª.
x.tofile('x.data.txt', '\n')
y.tofile('y.data.txt', '\n')

#–ß–∏—Ç–∞–µ–º –¥–∞–Ω–Ω—ã–µ –∏–∑ —Ñ–∞–π–ª–æ–≤.
x = fromfile('x.data.txt', float, sep='\n')
y = fromfile('y.data.txt', float, sep='\n')

print(x)
print(y)

#–ù–∞—Ö–æ–∂–¥–µ–Ω–∏–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ —Ñ—É–Ω–∫—Ü–∏–∏ –≤–∏–¥–∞ y = ax^2 + bx + c —Å –º–µ—Ç–æ–¥–æ–º –Ω–∞–∏–º–µ–Ω—å—à–∏—Ö –∫–≤–∞–¥—Ä–∞—Ç–æ–≤
#–ó–∞–¥–∞–µ–º –≤–µ–∫—Ç–æ—Ä m = [x**2, x, E]
m = vstack((x**2, x, ones(11))).T
#–ù–∞—Ö–æ–¥–∏–º –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –ø—Ä–∏ —Å–æ—Å—Ç–∞–≤–ª—è—é—â–∏—Ö –≤–µ–∫—Ç–æ—Ä–∞ m
s = np.linalg.lstsq(m, y, rcond=None)[0]

#–ù–∞ –æ—Ç—Ä–µ–∑–∫–µ [-5, 5]
x_prec = linspace(-5, 5, 101)
#–†–∏—Å—É–µ–º —Ç–æ—á–∫–∏
plt.plot(x, y, 'D')
#–†–∏—Å—É–µ–º –∫—Ä–∏–≤—É—é –≤–∏–¥–∞ y = ax^2 + bx + c, –ø–æ–¥—Å—Ç–∞–≤–ª—è—è –∏–∑ —Ä–µ—à–µ–Ω–∏—è –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã s[0], s[1], s[2]
plt.plot(x_prec, s[0] * x_prec**2 + s[1] * x_prec + s[2], '-', lw=2)
plt.grid()
plt.savefig('–ø–∞—Ä–∞–±–æ–ª–∞.png')
```

## 1.1.3 –ü—Ä–∏–º–µ—Ä
–ü–æ –¥–∞–Ω–Ω—ã–º –ø—Ä–µ–¥—ã–¥—É—â–µ–≥–æ –ø—Ä–∏–º–µ—Ä–∞ –ø–æ—Å—Ç—Ä–æ–π—Ç–µ —ç—Å—Ç—Ä–∞–ø–æ–ª—è—Ü–∏–æ–Ω–Ω–æ–≥–æ –ø–æ–ª–∏–Ω–æ–º–∞
—Ç—Ä–µ—Ç—å–µ–≥–æ –ø–æ—Ä—è–¥–∫–∞

```python
#–ù–∞—Ö–æ–∂–¥–µ–Ω–∏–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ —Ñ—É–Ω–∫—Ü–∏–∏ –≤–∏–¥–∞ y = ax^3 + bx^2 + cx + d —Å –º–µ—Ç–æ–¥–æ–º –Ω–∞–∏–º–µ–Ω—å—à–∏—Ö –∫–≤–∞–¥—Ä–∞—Ç–æ–≤
#–ó–∞–¥–∞–µ–º –≤–µ–∫—Ç–æ—Ä m = [x**3, x**2, x, E]
m = vstack((x**3, x**2, x, ones(11))).T
#–ù–∞—Ö–æ–¥–∏–º –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –ø—Ä–∏ —Å–æ—Å—Ç–∞–≤–ª—è—é—â–∏—Ö –≤–µ–∫—Ç–æ—Ä–∞ m
s = np.linalg.lstsq(m, y, rcond=None)[0]

#–ù–∞ –æ—Ç—Ä–µ–∑–∫–µ [-5, 5]
x_prec = linspace(-5, 5, 101)
#–†–∏—Å—É–µ–º —Ç–æ—á–∫–∏
plt.plot(x, y, 'D')
#–†–∏—Å—É–µ–º –∫—Ä–∏–≤—É—é –≤–∏–¥–∞ y = ax^3 + bx^2 + cx + d, –ø–æ–¥—Å—Ç–∞–≤–ª—è—è –∏–∑ —Ä–µ—à–µ–Ω–∏—è –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã s[0], s[1], s[2] s[3]
plt.plot(x_prec, s[0] * x_prec**3 + s[1] * x_prec**2 + s[2] * x_prec + s[3], '-', lw=3)
plt.grid()
plt.savefig('–ø–æ–ª–∏–Ω–æ–º –≤ 3-–π —Å—Ç–µ–ø–µ–Ω–∏.png')
```

–ü—Ä–µ–¥—Å—Ç–∞–≤—å—Ç–µ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –∏ –ø–æ—Å—Ç—Ä–æ–π—Ç–µ —ç–∫—Ç—Ä–∞–ø–æ–ª—è—Ü–∏—é –ø–æ–ª–∏–Ω–æ–º–∞–º–∏
–ø–µ—Ä–≤–æ–π, –≤—Ç–æ—Ä–æ–π –∏ —Ç—Ä–µ—Ç—å–µ–π —Å—Ç–µ–ø–µ–Ω–∏.

```python
x = linspace(-5, 5, 11)
y1 = x + rand(11)
y2 = x**2 + rand(11)

#–ü–æ–ª–∏–Ω–æ–º –ø–µ—Ä–≤–æ–π —Å—Ç–µ–ø–µ–Ω–∏
A = np.vstack((x, np.ones(len(x)))).T
m, c = np.linalg.lstsq(A, y1, rcond=None)[0]
x_prec = linspace(-5, 5, 101)

plt.plot(x, y1, 'D')
plt.plot(x_prec, m*x_prec + c, '-', lw=3)
plt.grid()
plt.show()

#–ü–æ–ª–∏–Ω–æ–º –≤—Ç–æ—Ä–æ–π —Å—Ç–µ–ø–µ–Ω–∏
m = np.vstack((x**2, x, np.ones(len(x)))).T
s = np.linalg.lstsq(m, y2, rcond=None)[0]
x_prec = linspace(-5, 5, 101)

plt.plot(x, y2, 'D')
plt.plot(x_prec, x_prec**2*s[0] + x_prec*s[1] + s[2], '-', lw=3)
plt.grid()
plt.show()

#–ü–æ–ª–∏–Ω–æ–º —Ç—Ä–µ—Ç—å–µ–π —Å—Ç–µ–ø–µ–Ω–∏
m = np.vstack((x**3, x**2, x, np.ones(len(x)))).T
s = np.linalg.lstsq(m, y2, rcond=None)[0]
x_prec = linspace(-5, 5, 101)

plt.plot(x, y2, 'D')
plt.plot(x_prec, x_prec**3*s[0] + x_prec**2*s[1] + x_prec*s[2] + s[3], '-', lw=3)
plt.grid()
plt.show()
```

## 1.1.4 –ü—Ä–∏–º–µ—Ä
–ù–µ–æ–±—Ö–æ–¥–∏–º–æ –ø—Ä–æ–≤–µ—Ä–∏—Ç—å –≥–∏–ø–æ—Ç–µ–∑—É, —á—Ç–æ –Ω–∞—à–∏ —Ç–æ—á–µ—á–Ω–æ –∑–∞–¥–∞–Ω–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –ª–æ–∂–∏—Ç—Å—è
–Ω–∞ –∫—Ä–∏–≤—É—é –≤–∏–¥–∞ ùëì(ùë•, ùëè) = ùëè0 + ùëè1ùëíùë•ùëù(‚àíùëè2ùë•
2
)

```python
#–î–æ–±–∞–≤–∏–º —à—É–º–∞ –≤ –¥–∞–Ω–Ω—ã–µ, —Å–¥–µ–ª–∞–Ω–Ω—ã–µ –ø–æ —Ñ—É–Ω–∫—Ü–∏–∏ f(x, b) —Å –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞–º–∏ b = (0.25, 0.75, 0.5)
beta = (0.25, 0.75, 0.5)
def f(x, b0, b1, b2):
    return b0 + b1 + np.exp(-b2*x**2)
#–ó–∞–¥–∞–¥–∏–º –º–∞—Å—Å–∏–≤ —Ç–æ—á–µ–∫ xi
xdata = np.linspace(0, 5, 50)
#–°–æ–∑–¥–∞–µ–º —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏ –ø—Ä–∞–≤–∏–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è —Ç–æ—á–µ–∫ yi (–±–µ–∑ —à—É–º–∞)
y = f(xdata, *beta)
#–ó–∞—à—É–º–ª—è–µ–º —ç—Ç–∏ –¥–∞–Ω–Ω—ã–µ
ydata = y + 0.05 * np.random.randn(len(xdata))
print(xdata)
print(ydata)
#–ò—Å–ø–æ–ª—å–∑—É–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è —Ä–µ—à–µ–Ω–∏—è –≤ –≤–∏–¥–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ —Ñ—É–Ω–∫—Ü–∏–∏ f(x) –¥–ª—è —É–∫–∞–∑–∞–Ω–Ω—ã—Ö xdata –∏ ydata
import scipy as sp
from scipy.optimize import curve_fit
beta_opt, beta_cov = sp.optimize.curve_fit(f, xdata, ydata)
print(beta_opt)
#–í—ã—á–∏—Å–ª–∏–º –ª–∏–Ω–µ–π–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ
lin_dev = sum(beta_cov[0])
print(lin_dev)

#–í—ã—á–∏—Å–ª–∏–º –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ
residuals = ydata - f(xdata, *beta_opt)
fres = sum(residuals**2)
print(fres)

fig, ax = plt.subplots()
ax.scatter(xdata, ydata)
ax.plot(xdata, y, 'r', lw=2)
ax.plot(xdata, f(xdata, *beta_opt), 'b', lw=2)
ax.set_xlim(0, 5)
ax.set_xlabel(r"$x$", fontsize=18)
ax.set_ylabel(r"$f(x, \beta)$", fontsize=18)
plt.show()
```

## 1.1.5 –ü—Ä–∏–º–µ—Ä
–ù–µ–æ–±—Ö–æ–¥–∏–º–æ –ø—Ä–æ–≤–µ—Ä–∏—Ç—å –≥–∏–ø–æ—Ç–µ–∑—É, —á—Ç–æ –Ω–∞—à–∏ —Ç–æ—á–µ—á–Ω–æ –∑–∞–¥–∞–Ω–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –ª–æ–∂–∏—Ç—Å—è
–Ω–∞ –∫—Ä–∏–≤—ã–µ –≤–∏–¥–∞:
1) ùëì(ùë•, ùëè) = ùëè0 + ùëè1ùë•
2) ùëì(ùë•, ùëè) = ùëè0 + ùëè1ùë• + ùëè2ùë•
2
3) ùëì(ùë•, ùëè) = ùëè0 + ùëè1ùëôùëõ(ùë•)
4) ùëì(ùë•, ùëè) = ùëè0 ùë•
ùëè1

```python
#—Ä–µ—à–µ–Ω–∏–µ 1
#–î–æ–±–∞–≤–∏–º —à—É–º–∞ –≤ –¥–∞–Ω–Ω—ã–µ, —Å–¥–µ–ª–∞–Ω–Ω—ã–µ –ø–æ —Ñ—É–Ω–∫—Ü–∏–∏ f(x, b) —Å –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞–º–∏ b = (0.25, 0.75)
beta = (0.25, 0.75)
def f(x, b0, b1):
    return b0 + b1*x
#–ó–∞–¥–∞–¥–∏–º –º–∞—Å—Å–∏–≤ —Ç–æ—á–µ–∫ xi
xdata = np.linspace(0, 5, 50)
#–°–æ–∑–¥–∞–µ–º —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏ –ø—Ä–∞–≤–∏–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è —Ç–æ—á–µ–∫ yi (–±–µ–∑ —à—É–º–∞)
y = f(xdata, *beta)
#–ó–∞—à—É–º–ª—è–µ–º —ç—Ç–∏ –¥–∞–Ω–Ω—ã–µ
ydata = y + 0.05*np.random.randn(len(xdata))
beta_opt, beta_cov = sp.optimize.curve_fit(f, xdata, ydata)
print(beta_opt)
#–í—ã—á–∏—Å–ª–∏–º –ª–∏–Ω–µ–π–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ
lin_dev = sum(beta_cov[0])
print(lin_dev)

#–í—ã—á–∏—Å–ª–∏–º –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ
residuals = ydata - f(xdata, *beta_opt)
fres = sum(residuals**2)
print(fres)

fig, ax = plt.subplots()
ax.scatter(xdata, ydata)
ax.plot(xdata, y, 'r', lw=2)
ax.plot(xdata, f(xdata, *beta_opt), 'b', lw=2)
ax.set_xlim(0, 5)
ax.set_xlabel(r"$x$", fontsize=18)
ax.set_ylabel(r"$f(x, \beta)$", fontsize=18)
plt.show()

#–†–µ—à–µ–Ω–∏–µ 2
#–î–æ–±–∞–≤–∏–º —à—É–º–∞ –≤ –¥–∞–Ω–Ω—ã–µ, —Å–¥–µ–ª–∞–Ω–Ω—ã–µ –ø–æ —Ñ—É–Ω–∫—Ü–∏–∏ f(x, b) —Å –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞–º–∏ b = (0.25, 0.75, 0.5)
beta = (0.25, 0.75, 0.5)
def f(x, b0, b1, b2):
    return b0 + b1*x + b2*x*x
#–ó–∞–¥–∞–¥–∏–º –º–∞—Å—Å–∏–≤ —Ç–æ—á–µ–∫ xi
xdata = np.linspace(0, 5, 50)
#–°–æ–∑–¥–∞–µ–º —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏ –ø—Ä–∞–≤–∏–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è —Ç–æ—á–µ–∫ yi (–±–µ–∑ —à—É–º–∞)
y = f(xdata, *beta)
#–ó–∞—à—É–º–ª—è–µ–º —ç—Ç–∏ –¥–∞–Ω–Ω—ã–µ
ydata = y + 0.05 * np.random.randn(len(xdata))
beta_opt, beta_cov = sp.optimize.curve_fit(f, xdata, ydata)
print(beta_opt)
#–í—ã—á–∏—Å–ª–∏–º –ª–∏–Ω–µ–π–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ
lin_dev = sum(beta_cov[0])
print(lin_dev)

#–í—ã—á–∏—Å–ª–∏–º –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ
residuals = ydata - f(xdata, *beta_opt)
fres = sum(residuals**2)
print(fres)

fig, ax = plt.subplots()
ax.scatter(xdata, ydata)
ax.plot(xdata, y, 'r', lw=2)
ax.plot(xdata, f(xdata, *beta_opt), 'b', lw=2)
ax.set_xlim(0, 5)
ax.set_xlabel(r"$x$", fontsize=18)
ax.set_ylabel(r"$f(x, \beta)$", fontsize=18)
plt.show()

#–†–µ—à–µ–Ω–∏–µ 3
#–î–æ–±–∞–≤–∏–º —à—É–º–∞ –≤ –¥–∞–Ω–Ω—ã–µ, —Å–¥–µ–ª–∞–Ω–Ω—ã–µ –ø–æ —Ñ—É–Ω–∫—Ü–∏–∏ f(x, b) —Å –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞–º–∏ b = (1, 2))
beta = (1, 2)
def f(x, b0, b1):
    return b0 + b1 * np.log(x)
#–ó–∞–¥–∞–¥–∏–º –º–∞—Å—Å–∏–≤ —Ç–æ—á–µ–∫ xi
xdata = np.linspace(1, 5, 50)
#–°–æ–∑–¥–∞–µ–º —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏ –ø—Ä–∞–≤–∏–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è —Ç–æ—á–µ–∫ yi (–±–µ–∑ —à—É–º–∞)
y = f(xdata, *beta)
#–ó–∞—à—É–º–ª—è–µ–º —ç—Ç–∏ –¥–∞–Ω–Ω—ã–µ
ydata = y + 0.05 * np.random.randn(len(xdata))
beta_opt, beta_cov = sp.optimize.curve_fit(f, xdata, ydata)
print(beta_opt)
#–í—ã—á–∏—Å–ª–∏–º –ª–∏–Ω–µ–π–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ
lin_dev = sum(beta_cov[0])
print(lin_dev)

#–í—ã—á–∏—Å–ª–∏–º –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ
residuals = ydata - f(xdata, *beta_opt)
fres = sum(residuals**2)
print(fres)

fig, ax = plt.subplots()
ax.scatter(xdata, ydata)
ax.plot(xdata, y, 'r', lw=2)
ax.plot(xdata, f(xdata, *beta_opt), 'b', lw=2)
ax.set_xlim(0, 5)
ax.set_xlabel(r"$x$", fontsize=18)
ax.set_ylabel(r"$f(x, \beta)$", fontsize=18)
plt.show()

#–†–µ—à–µ–Ω–∏–µ 4
#–î–æ–±–∞–≤–∏–º —à—É–º–∞ –≤ –¥–∞–Ω–Ω—ã–µ, —Å–¥–µ–ª–∞–Ω–Ω—ã–µ –ø–æ —Ñ—É–Ω–∫—Ü–∏–∏ f(x, b) —Å –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞–º–∏ b = (1, 2))
beta = (1, 2)
def f(x, b0, b1):
    return b0 * x ** b1
#–ó–∞–¥–∞–¥–∏–º –º–∞—Å—Å–∏–≤ —Ç–æ—á–µ–∫ xi
xdata = np.linspace(1, 5, 50)
#–°–æ–∑–¥–∞–µ–º —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏ –ø—Ä–∞–≤–∏–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è —Ç–æ—á–µ–∫ yi (–±–µ–∑ —à—É–º–∞)
y = f(xdata, *beta)
#–ó–∞—à—É–º–ª—è–µ–º —ç—Ç–∏ –¥–∞–Ω–Ω—ã–µ
ydata = y + 0.05 * np.random.randn(len(xdata))
beta_opt, beta_cov = sp.optimize.curve_fit(f, xdata, ydata)
print(beta_opt)
#–í—ã—á–∏—Å–ª–∏–º –ª–∏–Ω–µ–π–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ
lin_dev = sum(beta_cov[0])
print(lin_dev)

#–í—ã—á–∏—Å–ª–∏–º –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ
residuals = ydata - f(xdata, *beta_opt)
fres = sum(residuals**2)
print(fres)

fig, ax = plt.subplots()
ax.scatter(xdata, ydata)
ax.plot(xdata, y, 'r', lw=2)
ax.plot(xdata, f(xdata, *beta_opt), 'b', lw=2)
ax.set_xlim(0, 5)
ax.set_xlabel(r"$x$", fontsize=18)
ax.set_ylabel(r"$f(x, \beta)$", fontsize=18)
plt.show()
```

```python
#–ò–º–ø–æ—Ä—Ç–∏—Ä—É–µ–º –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã–µ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏
#–ò–º–ø–æ–ª—å–∑—É–µ–º pandas –∏ numpy –¥–ª—è –æ–±—Ä–∞–±–æ—Ç–∫–∏ –¥–∞–Ω–Ω—ã—Ö
#matplotlib –¥–ª—è –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏ –∏ sklearn –¥–ª—è –æ–±—É—á–µ–Ω–∏—è –Ω–∞–±–æ—Ä–æ–≤ –¥–∞–Ω–Ω—ã—Ö –∏ –∏–º–ø–æ—Ä—Ç–∞ –º–æ–¥–µ–ª–µ–π
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas import DataFrame, Series
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

#–°–æ–∑–¥–∞–¥–∏–º –Ω–∞–±–æ—Ä –¥–∞–Ω–Ω—ã—Ö –¥–ª—è –æ–ø–∏—Å–∞–Ω–∏—è –≤–∑–∞–∏–º–æ—Å–≤—è–∑–∏ –º–µ–∂–¥—É –≤—Ä–µ–º–µ–Ω–µ–º –æ–±—É—á–µ–Ω–∏—è —Å—Ç—É–¥–µ–Ω—Ç–æ–≤ –∏ —É—Å–ø–µ–≤–∞–µ–º–æ—Å—Ç—å—é
my_dict = {'–£—á–µ–±–Ω–æ–µ –≤—Ä–µ–º—è': [0.5, 0.75, 1, 1.25, 1.5, 1.75, 2, 2.25, 2.5, 2.75, 3, 3.25, 3.5, 3.75, 4, 4.25, 4.5, 4.75, 5, 5.5],
            '–û—Ü–µ–Ω–∫–∞': [10, 22, 13, 43, 20, 22, 33, 50, 62, 48, 55, 75, 62, 73, 81, 76, 64, 82, 90, 93]}

dataset = pd.DataFrame(my_dict)
dataset.head()
#–ò—Å—Å–ª–µ–¥—É–µ–º –Ω–∞–±–æ—Ä –¥–∞–Ω–Ω—ã—Ö
print(dataset.shape)
dataset.describe()
#–ù–∞—Ä–∏—Å—É–µ–º —Ç–æ—á–µ—á–Ω—É—é –¥–∏–∞–≥—Ä–∞–º–º—É
plt.scatter(dataset['–£—á–µ–±–Ω–æ–µ –≤—Ä–µ–º—è'], dataset['–û—Ü–µ–Ω–∫–∞'], color='b', label = '–¥–∞–Ω–Ω—ã–µ —ç–∫–∑–∞–º–µ–Ω–∞')
plt.xlabel('–ß–∞—Å—ã')
plt.ylabel('–û—Ü–µ–Ω–∫–∞')
plt.show()
```

–ù–∞–π–¥–∏—Ç–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –ª–∏–Ω–∏–∏ —Ä–µ–≥—Ä–µ—Å—Å–∏–∏. –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –ø—Ä–æ–≥–Ω–æ–∑.

```python
import pandas as pd
url = 'https://raw.githubusercontent.com/AnnaShestova/salary-years-simple-linear-regression/master/Salary_Data.csv'
dataset = pd.read_csv(url)

dataset.head()
#–ò—Å—Å–ª–µ–¥—É–µ–º –Ω–∞–±–æ—Ä –¥–∞–Ω–Ω—ã—Ö
print(dataset.shape)
dataset.describe()
#–ù–∞—Ä–∏—Å—É–µ–º —Ç–æ—á–µ—á–Ω—É—é –¥–∏–∞–≥—Ä–∞–º–º—É
plt.scatter(dataset['YearsExperience'], dataset['Salary'], color='b')
plt.xlabel('Years of experience')
plt.ylabel('Salary')
plt.show()

X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,1].values
print(X)
print(y)
#–¢–µ–ø–µ—Ä—å, –∫–æ–≥–¥–∞ —É –Ω–∞—Å –µ—Å—Ç—å –∞—Ç—Ä–∏–±—É—Ç—ã –∏ –º–µ—Ç–∫–∏, –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ —Ä–∞–∑–¥–µ–ª–∏—Ç—å –∏—Ö –Ω–∞ –æ–±—É—á–∞—é—â–∏–π –∏ —Ç–µ—Å—Ç–æ–≤—ã–π –Ω–∞–±–æ—Ä—ã.
#–ü—Ä–∏–≤–µ–¥–µ–Ω–Ω—ã–π —Ñ—Ä–∞–≥–º–µ–Ω—Ç —Ä–∞–∑–¥–µ–ª—è–µ—Ç 80% –¥–∞–Ω–Ω—ã—Ö –Ω–∞ –æ–±—É—á–∞—é—â–∏–π –Ω–∞–±–æ—Ä, –∞ 20% –¥–∞–Ω–Ω—ã—Ö - –Ω–∞ –Ω–∞–±–æ—Ä —Ç–µ—Å—Ç–æ–≤
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

#–î–∞–ª–µ–µ –º–æ–∂–Ω–æ –æ–±—É—á–∏—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
#–Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –∏–º–ø–æ—Ä—Ç–∏—Ä–æ–≤–∞—Ç—å –∫–ª–∞—Å—Å LinearRegression, —Å–æ–∑–¥–∞—Ç—å –µ–≥–æ —ç–∫–∑–µ–º–ø–ª—è—Ä –∏ –≤—ã–∑–≤–∞—Ç—å –º–µ—Ç–æ–¥ fit()
regressor = LinearRegression()
regressor.fit(X_train, y_train)
#–ü—Ä–∏–≤–µ–¥–µ–º –ø–æ–ª—É—á–∏–≤—à–∏–µ—Å—è –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –¥–ª—è –ª–∏–Ω–∏–∏ —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
print(regressor.intercept_)
print(regressor.coef_)

y_pred = regressor.predict(X_test)
# –°—Ä–∞–≤–Ω–∏–º —Ñ–∞–∫—Ç–∏—á–µ—Å–∫–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è —Å –ø—Ä–æ–≥–Ω–æ–∑–∏—Ä—É–µ–º—ã–º–∏
df = pd.DataFrame({'Actual': y_test, 'Predicted:': y_pred})
df
#–í–∏–∑—É–∞–ª–∏–∑–∏—Ä—É–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç —Å—Ä–∞–≤–Ω–µ–Ω–∏—è –≤ –≤–∏–¥–µ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã
df.plot(kind='bar')
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle='-', linewidth='0.5', color='black')
plt.show()
#–ü–æ—Å—Ç—Ä–æ–∏–º –ª–∏–Ω–∏—é —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ —Å —Ç–µ—Å—Ç–æ–≤—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏
plt.scatter(X_test, y_test, color='gray')
plt.plot(X_test, y_pred, color='red', linewidth=2)
plt.show()
```

```python
import numpy as np

y = [1,2,3,4,3,4,5,3,5,5,4,5,4,5,4,5,6,0,6,3,1,3,1]
X = [[0,2,4,1,5,4,5,9,9,9,3,7,8,8,6,6,5,5,5,6,6,5,5],
     [4,1,2,3,4,5,6,7,5,8,7,8,7,8,7,8,6,8,9,2,1,5,6],
     [4,1,2,5,6,7,8,9,7,8,7,8,7,4,3,1,2,3,4,1,3,9,7]]
X = np.transpose(X) # transpose to input vectors
X = np.c_[X, np.ones(X.shape[0])] # add bias term
linreg = np.linalg.lstsq(X, y, rcond=None)[0]
print(linreg)
```

## 1.3.2 –ü—Ä–∏–º–µ—Ä
–î–ª—è –¥–∞–Ω–Ω—ã—Ö –∏–∑ –ø—Ä–µ–¥—ã–¥—É—â–µ–π –∑–∞–¥–∞—á–∏ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –º–æ–¥–µ–ª—å –º–Ω–æ–∂–µ—Å—Ç–≤–µ–Ω–Ω–æ–π
–ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º —Å—Ä–µ–¥—Å—Ç–≤ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ s—Åikit-learn.

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

#–ò–º–ø–æ—Ä—Ç–∏—Ä—É–µ–º –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã–µ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as seabornInstance
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

y = [1,2,3,4,3,4,5,3,5,5,4,5,4,5,4,5,6,0,6,3,1,3,1]
X = [[0,2,4,1,5,4,5,9,9,9,3,7,8,8,6,6,5,5,5,6,6,5,5],
     [4,1,2,3,4,5,6,7,5,8,7,8,7,8,7,8,6,8,9,2,1,5,6],
     [4,1,2,5,6,7,8,9,7,8,7,8,7,4,3,1,2,3,4,1,3,9,7]]

# –§–æ—Ä–º–∏—Ä—É–µ–º DataFrame –∏–∑ –¥–≤—É—Ö —Å–ø–∏—Å–∫–æ–≤
new_y = np.array(y)
new_y = new_y.transpose()
df1 = pd.DataFrame(new_y)
new_X = np.array(X)
new_X = new_X.transpose()
df2 = pd.DataFrame(new_X)
df1 = df1.rename(columns = {0: 'y'}, inplace = False)
df2 = df2.rename(columns = {0: 'x1', 1: 'x2', 2: 'x3'}, inplace = False)

frames = [df1, df2]
dataset = pd.concat([df1, df2], axis=1, join='inner')
dataset.head()

#–∏–∑—É—á–∏–º –¥–∞–Ω–Ω—ã–µ
print(dataset.shape)
dataset.describe()

#—Ä–∞–∑–¥–µ–ª–∏–º –¥–∞–Ω–Ω—ã–µ –Ω–∞ –º–µ—Ç–∫–∏ –∏ –∞—Ç—Ä–∏–±—É—Ç—ã
X = dataset[['x1', 'x2', 'x3']]
y = dataset['y']

#—Ä–∞–∑–¥–µ–ª–∏–º –¥–∞–Ω–Ω—ã–µ –Ω–∞ –æ–±—É—á–∞—é—â–∏—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫–∏
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

#–¥–ª—è –æ–±—É—á–µ–Ω–∏—è –∞–ª–≥–æ—Ä–∏—Ç–º–∞ –º—ã –≤—ã–ø–æ–ª–Ω—è–µ–º —Ç–æ—Ç –∂–µ –∫–æ–¥, —á—Ç–æ –∏ —Ä–∞–Ω—å—à–µ, –∏—Å–ø–æ–ª—å–∑—É—è –º–µ—Ç–æ–¥ fit() –∫–ª–∞—Å—Å–∞ linearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)

#–≤—ã–±–µ—Ä–µ–º –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –º–æ–¥–µ–ª–∏
coeff_df = pd.DataFrame(regressor.coef_, X.columns, columns=['Coefficient'])
coeff_df

#–ß—Ç–æ–±—ã —Å–¥–µ–ª–∞—Ç—å –ø—Ä–æ–≥–Ω–æ–∑—ã –Ω–∞ —Ç–µ—Å—Ç–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö, –≤—ã–ø–æ–ª–Ω–∏—Ç–µ —Å–ª–µ–¥—É—é—â–∏–π –∫–æ–¥
y_pred = regressor.predict(X_test)
df = pd.DataFrame({'Actual':y_test, 'Predicted':y_pred})
df

#–ü–æ—Å–ª–µ–¥–Ω–∏–π —à–∞–≥ - –æ—Ü–µ–Ω–∏—Ç—å –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º–∞. –ú—ã —Å–¥–µ–ª–∞–µ–º —ç—Ç–æ, –Ω–∞–π–¥—è –∑–Ω–∞—á–µ–Ω–∏—è –¥–ª—è MSE
print('Mean Squared Error: ', metrics.mean_squared_error(y_test, y_pred))
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

url1 = 'https://raw.githubusercontent.com/likarajo/petrol_consumption/master/data/petrol_consumption.csv'
url2 = 'https://raw.githubusercontent.com/aniruddhachoudhury/Red-Wine-Quality/master/winequality-red.csv'

dataset1 = pd.read_csv(url1)
dataset2 = pd.read_csv(url2)

X_train1, X_test1, y_train1, y_test1 = train_test_split(dataset1.iloc[:, :-1],
                                                        dataset1.iloc[:, -1],
                                                        test_size=0.2,
                                                        random_state=0)

X_train2, X_test2, y_train2, y_test2 = train_test_split(dataset2.iloc[:, :-1],
                                                        dataset2.iloc[:, -1],
                                                        test_size=0.2,
                                                        random_state=0)

regressor1 = LinearRegression()
regressor1.fit(X_train1, y_train1)

regressor2 = LinearRegression()
regressor2.fit(X_train2, y_train2)

y_pred1 = regressor1.predict(X_test1)
df1 = pd.DataFrame({'Actual': y_test1, 'Predicted': y_pred1})
display(df1)

y_pred2 = regressor2.predict(X_test2)
df2 = pd.DataFrame({'Actual': y_test2, 'Predicted': y_pred2})
display(df2)

print('Mean Squared Error first csv: ', metrics.mean_squared_error(y_test1, y_pred1))
print('Mean Squared Error second csv: ', metrics.mean_squared_error(y_test2, y_pred2))
```

## –ó–∞–¥–∞–Ω–∏–µ*

—Å—Ç–µ–ø–µ–Ω–∏, –∞–ø–ø—Ä–æ–∫—Å–∏–º–∏—Ä—É—é—â–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞, —Å –ø—Ä–∏–º–µ–Ω–µ–Ω–∏–µ–º
–º–µ—Ç–æ–¥–∞ –Ω–∞–∏–º–µ–Ω—å—à–∏—Ö –∫–≤–∞–¥—Ä–∞—Ç–æ–≤. –†–µ–∑—É–ª—å—Ç–∞—Ç—ã –≤—ã–≤–æ–¥—è—Ç—Å—è –≤ –≤–∏–¥–µ —Ç–∞–±–ª–∏—Ü
–∑–Ω–∞—á–µ–Ω–∏–π –∏ –≥—Ä–∞—Ñ–∏–∫–æ–≤, –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö –ø–æ–ª–∏–Ω–æ–º–æ–≤

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([4.0, 4.2, 4.4, 4.6, 4.8, 5.0])
y = np.array([4.0, 3.0, 6.0, 6.0, 4.0, 4.0])

A = np.vstack([x, np.ones(len(x))]).T
print(A)

m, c = np.linalg.lstsq(A, y, rcond = None)[0]
print(m, c)

plt.plot(x, y, 'o', label = '–ò—Å—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ', markersize = 10)
plt.plot(x, m*x + c, 'r', label = '–õ–∏–Ω–µ–π–Ω–∞—è —ç–∫—Å—Ç—Ä–∞–ø–æ–ª—è—Ü–∏—è')
plt.legend()
plt.show()

m = np.vstack((x ** 2, x, np.ones(6))).T

s = np.linalg.lstsq(m, y, rcond = None)[0]

x_prec = np.linspace(4, 5, 6)

plt.plot(x, y, 'D')
plt.plot(x_prec, s[0] * x_prec ** 2 + s[1] * x_prec + s[2], '-', lw = 2)
plt.grid()
print(s)
```
