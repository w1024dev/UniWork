```python
import pandas as pd

def credit_calc(credit_sum, credit_period, rate):
    months = ["Январь", "Февраль", "Март", "Апрель", "Май", "Июнь", "Июль", "Август", "Сентябрь", "Октябрь", "Ноябрь", "Декабрь"]
    df = pd.DataFrame(columns=["Период", "Задолженность на начало периода", "Ежемесячный платеж", "Остаток по задолженности"])
    
    credit_period *= 12 # из лет в месяцы
    rate /= 12
    whole_sum = credit_sum*(credit_period*rate/(1-(1/(1+rate))**credit_period))
    print("Paid total:", whole_sum)
    monthly_payment = whole_sum/credit_period
    i = 0;
    
    while i < credit_period:
        old_sum = credit_sum
        if(monthly_payment > (1+rate)*credit_sum):
            monthly_payment = (1+rate)*credit_sum
        credit_sum = (1+rate)*credit_sum - monthly_payment
        df.loc[len(df)] = [months[i%12] + "_Год_" + str(i//12+1), old_sum, monthly_payment, credit_sum]
        i += 1
    return df
```

```python
summ = int(input("Input loan size: "))
period = int(input("Input loan period: "))
rate = float(input("Input loan rate: "))
temp = credit_calc(summ, period, rate)
temp = temp.applymap(lambda x: round(x, 2) if isinstance(x, (int, float)) else x)
pd.set_option('display.max_rows', None)
print(temp)
```
