```python
import math
import time

pi = math.pi
```

### Решить систему нелинейных уравнений методом итерации с точностью до 0.001 и до 0.0001.


```python
def x_new(y_old):
    return (-0.4 - math.sin(y_old))

def y_new(x_old):
    return (math.cos(x_old + 1)/2)

def recursive_function(x_old, y_old, accuracy, iteration = 1):
    print(f"Iteration {iteration}")
    print(f"[ x,  y]: {x_new(y_old):.5f}, {y_new(x_old):.5f}")
    
    dx = abs(x_old-x_new(y_old))
    dy = abs(y_old-y_new(x_old))
    print(f"[dx, dy]: {dx:.5f}, {dy:.5f}")
    
    if max(dx, dy) < accuracy:
        return [x_new(y_old), y_new(x_old), iteration]
    else:
        return recursive_function(x_new(y_old), y_new(x_old), accuracy, iteration+1)


print("Accuracy = 0.001")

start = time.perf_counter()
iteration_method_res_001 = recursive_function(0, 0, 0.001)
end = time.perf_counter()
iteration_method_res_001.append(end-start)
iteration_method_res_001.append(0.001)

print(f"Final values for x, y:\n{iteration_method_res_001[:2]}")
print(f"Iterations: {iteration_method_res_001[2]}")
print(f"Time taken: {iteration_method_res_001[3]:.5f} seconds")

print("\nAccuracy = 0.0001")

start = time.perf_counter()
iteration_method_res_0001 = recursive_function(0, 0, 0.0001)
end = time.perf_counter()
iteration_method_res_0001.append(end-start)
iteration_method_res_0001.append(0.0001)

print(f"Final values for x, y:\n{iteration_method_res_0001[:2]}")
print(f"Iterations: {iteration_method_res_0001[2]}")
print(f"Time taken: {iteration_method_res_0001[3]:.5f} seconds")
```

### Решить систему нелинейных уравнений методом Ньютона с точностью до 0.001 и до 0.0001.


```python
# J = (math.sin(x+1), 2)
#     (1,   math.cos(y))

# Умножение матрицы Якоби на dx,dy
# math.sin(x_old+1)*dx + 2*dy = -f1
# 1*dx + math.cos(y_old)*dy = -f2

# Выражение dy
# 2*dy = -f1 - math.sin(x_old+1)*dx
# 2*dy = -f1 - math.sin(x_old+1)*(-f2 - math.cos(y_old)*dy)
# 2*dy = -f1 + f2*math.sin(x_old+1) + math.sin(x_old+1)*math.cos(y_old)*dy
# dy*(2 - math.sin(x_old+1)*math.cos(y_old)) = -f1 + f2*math.sin(x_old+1)
# dy = (-f1 + math.sin(x_old+1)*f2)/(2 - math.sin(x_old+1)*math.cos(y_old))

def recursive_function(x_old, y_old, accuracy, iteration = 1):
    print(f"Iteration {iteration}")
    
    f1 = 2*y_old - math.cos(x_old+1)
    f2 = x_old + math.sin(y_old) + 0.4
    
    dy = (-1 * f1 + math.sin(x_old+1)*f2)/(2 - math.sin(x_old+1)*math.cos(y_old))
    dx = -1 * f2 - math.cos(y_old)*dy
    
    print(f"[ x,  y]: {x_old+dx:.5f}, {y_old+dy:.5f}")
    print(f"[dx, dy]: {dx:.5f}, {dy:.5f}")
    
    if max(dx, dy) < accuracy:
        return [x_old+dx, y_old+dy, iteration]
    else:
        return recursive_function(x_old+dx, y_old+dy, accuracy, iteration+1)
    

print("Accuracy = 0.001")

start = time.perf_counter()
newton_method_result_001 = recursive_function(0, 0, 0.001)
end = time.perf_counter()
newton_method_result_001.append(end-start)
newton_method_result_001.append(0.001)

print(f"Final values for x, y:\n{newton_method_result_001[:2]}")
print(f"Iterations: {newton_method_result_001[2]}")
print(f"Time taken: {newton_method_result_001[3]:.5f} seconds")

print("\nAccuracy = 0.0001")

start = time.perf_counter()
newton_method_result_0001 = recursive_function(0, 0, 0.0001)
end = time.perf_counter()
newton_method_result_0001.append(end-start)
newton_method_result_0001.append(0.0001)

print(f"Final values for x, y:\n{newton_method_result_0001[:2]}")
print(f"Iterations: {newton_method_result_0001[2]}")
print(f"Time taken: {newton_method_result_0001[3]:.5f} seconds")
```

### Решить систему нелинейных уравнений модифицированным методом Ньютона с точностью до 0.001 и 0.0001.


```python
def recursive_function(x_old, y_old, J, accuracy, iteration = 1):
    print(f"Iteration {iteration}")
    
    f1 = 2*y_old - math.cos(x_old+1)
    f2 = x_old + math.sin(y_old) + 0.4
    
    dy = (-1 * f1 + J[0]*f2)/(2 - J[0]*J[3])
    dx = -1 * f2 - J[3]*dy
    
    print(f"[ x,  y]: {x_old+dx:.5f}, {y_old+dy:.5f}")
    print(f"[dx, dy]: {dx:.5f}, {dy:.5f}")
    
    if max(dx, dy) < accuracy:
        return [x_old+dx, y_old+dy, iteration]
    else:
        return recursive_function(x_old+dx, y_old+dy, J, accuracy, iteration+1)
        
def Jacobian(x, y):
    return [math.sin(x+1), 2, 
            1, math.cos(y)]


J = Jacobian(0, 0)
print("Accuracy = 0.001")

start = time.perf_counter()
modified_newton_method_result_001 = recursive_function(0, 0, J, 0.001)
end = time.perf_counter()
modified_newton_method_result_001.append(end-start)
modified_newton_method_result_001.append(0.001)

print(f"Final values for x, y:\n{modified_newton_method_result_001[:2]}")
print(f"Iterations: {modified_newton_method_result_001[2]}")
print(f"Time taken: {modified_newton_method_result_001[3]:.5f} seconds")

print("\nAccuracy = 0.0001")

start = time.perf_counter()
modified_newton_method_result_0001 = recursive_function(0, 0, J, 0.0001)
end = time.perf_counter()
modified_newton_method_result_0001.append(end-start)
modified_newton_method_result_0001.append(0.0001)

print(f"Final values for x, y:\n{modified_newton_method_result_0001[:2]}")
print(f"Iterations: {modified_newton_method_result_0001[2]}")
print(f"Time taken: {modified_newton_method_result_0001[3]:.5f} seconds")
```
